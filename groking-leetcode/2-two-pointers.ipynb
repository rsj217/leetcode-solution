```python
from typing import *

class Solution:
    def reverseString(self, s: List[str]) -> None:
        l = 0
        r = len(s) - 1
        while l < r:
            s[l], s[r] = s[r], s[l]
            l += 1
            r -= 1
```

ä»€ä¹ˆæ—¶å€™ç§»åŠ¨å‘¢ï¼Ÿç”± [l, r] ç»„æˆçš„çª—å£ï¼Œå…¶ç››æ°´é‡ä¸º `min(l, r) * (r - l)`ã€‚ç”±æ­¤å¯è§ï¼Œr-l åªèƒ½æ›´å°ï¼Œæƒ³è¦è¿™ä¸ªè¡¨è¾¾å¼æ›´å¤§ï¼Œåªèƒ½æ‰¾åˆ°æ›´é«˜çš„æ•°ã€‚å› æ­¤æƒ³è¦ç››æ›´å¤šçš„æ°´ï¼Œåªæœ‰æŠŠæœ€çŸ®çš„è®¾æ³•å˜é«˜ï¼Œå³é«˜åº¦æ›´çŸ®çš„æŒ‡é’ˆè¿›è¡Œç§»åŠ¨ã€‚å¦‚æœç§»åŠ¨æ›´é«˜çš„æŒ‡é’ˆï¼Œé‚£ä¹ˆ r - l è·ç¦»å˜å°äº†ï¼ŒåŒæ—¶ `min(l, r)` ä¹Ÿä¸ä¼šæ›´å¤§ï¼Œåªä¼šæ›´å°ã€‚æœ€ç»ˆä»£ç å¦‚ä¸‹ï¼š

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        l = 0
        r = len(height) - 1
        max_area = float('-inf')
        while l < r:
            min_height = min(height[l], height[r])
            max_area = max(min_height * (r - l), max_area)
            if height[l] < height[r]:
                l += 1
            else:
                r -= 1
        return max_area if max_area != float('-inf') else 0 
```

ä¸€ä¸ªç®€å•åŠæ³•å°±æ˜¯ä½¿ç”¨å¯¹æ’æŒ‡é’ˆï¼Œl å’Œ r åˆ†åˆ«å‘ä¸­é—´é æ‹¢ï¼Œç§»åŠ¨çš„è¿‡ç¨‹ä¸­å¯ä»¥æ‰¾å‡º [0, l] ä¸­é—´çš„æœ€é«˜ l_max å’Œ [r, len]ä¸­é—´çš„æœ€é«˜ r_maxï¼Œè€Œå½“å‰çš„æŒ‡é’ˆå¯ä»¥å– min(l_max, r_max) ä¸­çš„éƒ¨åˆ†ã€‚å¦‚ä¸‹å›¾

å½“ l , r åˆ†åˆ«åœ¨ä¸Šå›¾çš„è¿‡ç¨‹ä¸­ï¼Œl æ‰€èƒ½ç››æ°´çš„å®¹é‡æ˜¯ ` height[l_max] - height[l]`ï¼Œæ­¤æ—¶æ˜¯0ï¼Œç„¶å l å‘å³ç§»åŠ¨ã€‚
æ­¤æ—¶ l èƒ½ç››æ°´ ä¾ç„¶æ˜¯  ` height[l_max] - height[l]` åˆ™æ˜¯ä¸€ä¸ªå•ä½ã€‚ä¾æ¬¡ç±»æ¨ï¼Œåªè¦ l_max < r_max ï¼Œé‚£ä¹ˆå°±ç§»åŠ¨ l æŒ‡é’ˆï¼Œç›´åˆ° l ç­‰äº rã€‚å¦‚æœ l_max å°äº r_maxï¼Œä¹Ÿå°±æ˜¯ä¸Šå›¾çš„å¯¹ç§°è¿‡ç¨‹ï¼Œå°±ä¸å†èµ˜è¿°ã€‚å…·ä½“å¯ä»¥çœ‹ä»£ç ï¼š

```python
class Solution:
    def trap(self, height: List[int]) -> int:
        if len(height) <= 1:
            return 0
        l = 0
        r = len(height) - 1
        l_max = height[l]
        r_max = height[r]

        ret = 0
        while l <= r:
            l_max = max(l_max, height[l])  # æ‰¾å‡ºå·¦è¾¹æœ€å¤§çš„æŸ±å­
            r_max = max(r_max, height[r]) # æ‰¾å‡ºå³è¾¹æœ€å¤§çš„æŸ±å­
            
            if l_max < r_max:                       
                ret += l_max - height[l]         # ç§»åŠ¨å·¦è¾¹
                l += 1
            else:
                ret += r_max - height[r]
                r -= 1
        return ret
```

æœ‰åºçš„æ•°ç»„å°±æ¯”å¥½åŠäº†ï¼Œå¯¹äºä¸¤ä¸ªæ•°ä¹‹å’Œï¼Œå¦‚ä½•å¤§äºtargetï¼Œé‚£ä¹ˆå°±éœ€è¦å‡å°‘ä¸¤æ•°ä¹‹å’Œï¼Œè‡ªç„¶å°±æ˜¯ r æŒ‡é’ˆå‘å·¦ç§»åŠ¨ï¼Œå¦‚æœå°äºtargetå…ƒç´ ï¼Œé‚£ä¹ˆå°±éœ€è¦å¢åŠ ä¸¤æ•°ä¹‹å’Œï¼Œè‡ªç„¶å°±æ˜¯ç§»åŠ¨å·¦è¾¹çš„ræŒ‡é’ˆã€‚

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        if len(numbers) < 2:
            return [-1, -1]
        l = 0
        r = len(numbers) - 1
        while l < r:                                         
            if numbers[l] + numbers[r] < target:
                l += 1
            elif target < numbers[l] + numbers[r]:
                r -= 1
            else:
                return [l+1, r+1]
        return [-1, -1]
```

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        ret = []
        for i in range(len(nums)-2):
            if i > 0 and nums[i-1] == nums[i]:
                continue
 
            l = i + 1
            r = len(nums) - 1
                   
            while l < r:
                if nums[i] + nums[l] + nums[r] < 0:
                    l += 1
                elif 0 < nums[i] + nums[l] + nums[r]:
                    r -= 1
                else:
                    ret.append([nums[i], nums[l], nums[r]])
                    l += 1
                    r -= 1
                    
                    while l < r and nums[l-1] == nums[l]:
                        l += 1

                    while l < r and nums[r] == nums[r+1]:
                        r -= 1
                    continue
        return ret
```

```python
class Solution:
    def threeSumClosest(self, nums: List[int], target: int) -> int:
        if len(nums) < 3:
            return None
        
        nums.sort()
        min_distance = float('inf')
        ret = None
        for i in range(len(nums) - 2):
            l = i + 1
            r = len(nums) - 1
            
            while l < r:
                three_sum = nums[i] + nums[l] + nums[r]
                distance = three_sum - target
                
                if abs(distance) < min_distance:  # å¦‚æœå‡ºç°æ›´çŸ­çš„è·ç¦»ï¼Œå³æ›´æ¥è¿‘targetï¼Œåˆ™æ›´æ–°è§£
                    min_distance = abs(distance)
                    ret = three_sum
                
                if 0 < distance:                             # è·ç¦»å¤§äº 0ï¼Œå³è¡¨ç¤ºå’Œæ›´å¤§ï¼Œå¯ä»¥å‡å°å’Œï¼Œå³ r æŒ‡é’ˆå·¦ç§»
                    r -= 1
                elif distance < 0 :                         # è·ç¦»å°äº 0ï¼Œå³è¡¨ç¤ºå’Œæ›´å°ï¼Œå¯ä»¥å¢åŠ å’Œï¼Œå³ l æŒ‡é’ˆå³ç§»
                    l += 1
                else:
                    return three_sum
        return ret
```

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        l = -1
        r = len(nums)
        i = 0
        while i < r:
            if nums[i] == 0:
                nums[l+1], nums[i] = nums[i], nums[l+1]  # ä»å·¦å¾€å³æ‰«æï¼Œå› ä¸ºäº¤æ¢åçš„ å…ƒç´  l+1 å·²ç»æ‰«æäº†ï¼Œå› æ­¤éœ€è¦ i += 1
                l += 1
                i += 1
            elif nums[i] == 1:
                i += 1
            else:
                nums[r-1], nums[i] = nums[i], nums[r-1] # äº¤æ¢åçš„å…ƒç´ è¿˜æ²¡ä¸çŸ¥é“æ˜¯å¤šå°‘ï¼Œéœ€è¦é’ˆå¯¹è¿™ä¸ªè¿›è¡Œå¤„ç†
                r -= 1
```

é’ˆå¯¹è¿™é¢˜ï¼Œæ—¢è¦é…åˆæ»‘åŠ¨çª—å£ï¼Œï¼Œä¹Ÿè¦åˆç†çš„ä½¿ç”¨å¯¹æ’æŒ‡é’ˆã€‚ä½¿ç”¨æ»‘åŠ¨çª—å£ç¡®å®šä¸€ä¸ªå¯èƒ½è§£ï¼Œç”±äºä¹˜ç§¯æœ€å°ï¼Œé‚£ä¹ˆçª—å£æœ€å³è¾¹çš„å…ƒç´ å¿…é¡»åŒ…å«ï¼Œé‚£ä¹ˆå†æ±‚è§£è¿™ä¸ªçª—å£åŒ…å«æœ€æœ‰å…ƒç´ çš„å­æ•°ç»„ï¼Œå°±æ˜¯çª—å£å†…çš„æ‰€æœ‰è§£ã€‚å…·ä½“ä»£ç å¦‚ä¸‹ï¼š

```python
class Solution:
    def numSubarrayProductLessThanK(self, nums: List[int], k: int) -> int:
        ret = 0
        lo = hi = 0
        product = 1
        while hi < len(nums):
            product *= nums[hi]
            hi += 1
            while product >= k and lo < hi:
                product /= nums[lo]
                lo += 1
            ret += hi - lo
        return ret
```
